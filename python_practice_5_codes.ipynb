```python
# 1. Simple Calculator
a = 10
b = 5
print('Addition:', a + b)
print('Subtraction:', a - b)
print('Multiplication:', a * b)
print('Division:', a / b)
```

```python
# 2. Check Even or Odd
num = 7
if num % 2 == 0:
    print('Even Number')
else:
    print('Odd Number')
```

```python
# 3. Factorial using loop
n = 5
fact = 1
for i in range(1, n+1):
    fact *= i
print('Factorial of', n, 'is', fact)
```

```python
# 4. List and Loop Example
fruits = ['Apple', 'Banana', 'Mango']
for fruit in fruits:
    print(fruit)
```

```python
# 5. Simple Function Example
def greet(name):
    return f'Hello, {name}!'

print(greet('Diya'))
```
